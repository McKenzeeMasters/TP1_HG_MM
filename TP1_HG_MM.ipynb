```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from prettytable import PrettyTable as pt
```

```python
Super=pd.read_csv('Superstore.csv', encoding='windows-1252')
Super.head()
```

```python
Super.info()
```

### Removing spaces in the columns 

```python
Super.columns =[column.replace(" ", "_")  for column in Super.columns]
Super.head()
```

### Converting the dates from object to data time format

```python
Super['Order_Date']=pd.to_datetime(Super['Order_Date'])
Super['Ship_Date']=pd.to_datetime(Super['Ship_Date'])
```

```python
Super.head()
```

```python
Super.info()
```

### Adding columns Order_Year, Order_Month, Ship_Year, and Ship_Month

```python
Super['Order_Year']=Super['Order_Date'].dt.year
Super['Order_Month']=Super['Order_Date'].dt.month
Super['Ship_Year']=Super['Ship_Date'].dt.year
Super['Ship_Month']=Super['Ship_Date'].dt.month
```

### Sales Plots

```python
#Sales distibution
plt.figure(figsize=(10,8))

#mean and median
mean=Super['Sales'].mean()
median=Super['Sales'].median()

sns.histplot(x='Sales', 
            data=Super)

plt.axvline(mean, color='yellow')
plt.axvline(median, color='white')

plt.show()
```

```python
plt.figure(figsize=(20,16))
sns.boxplot(x=Super['Sales'])

plt.show()
```

```python
#line plot of sales over the years
sy=sns.lineplot(x='Order_Year',
                y='Sales', 
                data=Super)
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales from 2014-2017')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(10,8)})
bar_plot=sns.barplot(x='Order_Year',
                    y='Sales', 
                    data=Super,
                    ci=False,
                    hue='State'
                    )
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales by State (2014-2017)')
plt.legend(bbox_to_anchor=(1.02, 1.5), loc='upper left')

plt.show()
```
